# Loading the dataset

```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
url = "seafood-and-fish-production-thousand-tonnes (1).csv"
seafood_df = pd.read_csv(url)

# Displaying the first few rows of the dataset
seafood_df.head()
```

# Chart 1: Line plot showing the trends of fish production over the years for different categories

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chart 1: Line plot showing the trends of fish production over the years for different categories
plt.figure(figsize=(12, 6))
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes', label='Pelagic Fish')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Crustaceans - 2765 - Production - 5510 - tonnes', label='Crustaceans')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Cephalopods - 2766 - Production - 5510 - tonnes', label='Cephalopods')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Demersal Fish - 2762 - Production - 5510 - tonnes', label='Demersal Fish')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Freshwater Fish - 2761 - Production - 5510 - tonnes', label='Freshwater Fish')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Molluscs, Other - 2767 - Production - 5510 - tonnes', label='Molluscs, Other')
sns.lineplot(data=seafood_df, x='Year', y='Commodity Balances - Livestock and Fish Primary Equivalent - Marine Fish, Other - 2764 - Production - 5510 - tonnes', label='Marine Fish, Other')
plt.title('Fish Production Trends Over Years for Different Categories')
plt.xlabel('Year')
plt.ylabel('Fish Production (Thousand Tonnes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Chart 2: Box plot to visualize the distribution of fish production for each category

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(data=seafood_df.drop(['Year'], axis=1))
plt.title('Distribution of Fish Production for Each Category')
plt.xlabel('Category')
plt.ylabel('Fish Production (Thousand Tonnes)')
plt.xticks(rotation=45)
plt.show()
```

# Chart 3: Bar plot showing the total fish production by year

```python
import matplotlib.pyplot as plt

# Ensure only numeric columns are summed
numeric_columns = seafood_df.select_dtypes(include='number').columns
total_production_by_year = seafood_df.groupby('Year')[numeric_columns].sum().sum(axis=1)

plt.figure(figsize=(12, 6))
total_production_by_year.plot(kind='bar', color='green')
plt.title('Total Fish Production by Year')
plt.xlabel('Year')
plt.ylabel('Total Fish Production (Thousand Tonnes)')
plt.xticks(rotation=45)
plt.show()
```

# Chart 4: Bar plot showing the total fish production by country

```python
import matplotlib.pyplot as plt

total_production_by_country = seafood_df.groupby('Entity')['Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
total_production_by_country.plot(kind='bar', color='blue')
plt.title('Total Fish Production by Country')
plt.xlabel('Country')
plt.ylabel('Total Fish Production (Thousand Tonnes)')
plt.xticks(rotation=90)
plt.show()
```

# Trends in Fish Production

```python
# Example: Trends in fish production for a specific country
country_name = 'United States'
country_production = seafood_df[seafood_df['Entity'] == country_name].groupby('Year')['Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes'].sum()
print(country_production)
```

#  Comparative Analysis

```python
# Example: Comparative analysis of fish production between two countries
country1 = 'United States'
country2 = 'China'
production_country1 = seafood_df[seafood_df['Entity'] == country1].groupby('Year')['Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes'].sum()
production_country2 = seafood_df[seafood_df['Entity'] == country2].groupby('Year')['Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes'].sum()
plt.figure(figsize=(12, 6))
plt.plot(production_country1, label=country1)
plt.plot(production_country2, label=country2)
plt.title('Comparative Analysis of Fish Production')
plt.xlabel('Year')
plt.ylabel('Fish Production (Thousand Tonnes)')
plt.legend()
plt.show()
```

# Seasonal Variations

```python
# Example: Seasonal analysis of fish production for a specific country
country_name = 'India'
seasonal_production = seafood_df[seafood_df['Entity'] == country_name].groupby('Year')['Commodity Balances - Livestock and Fish Primary Equivalent - Pelagic Fish - 2763 - Production - 5510 - tonnes'].mean()
# Plotting seasonal variations
plt.figure(figsize=(12, 6))
seasonal_production.plot(kind='bar')
plt.title('Seasonal Variations in Fish Production (Pelagic Fish) - ' + country_name)
plt.xlabel('Year')
plt.ylabel('Average Fish Production (Thousand Tonnes)')
plt.xticks(rotation=45)
plt.show()
```

# Identifying Overfished Species

```python
# Calculate the total production by summing production values across all fish categories
seafood_df['Total_Production'] = seafood_df.iloc[:, 3:].sum(axis=1)

# Calculate the threshold value based on the mean or median of total production
threshold_value = seafood_df['Total_Production'].mean()  # Adjust this according to your criteria

# Identify overfished species
total_production_by_species = seafood_df.groupby('Entity')['Total_Production'].sum().sort_values(ascending=False)
overfished_species = total_production_by_species[total_production_by_species < threshold_value]

# Print overfished species
print(overfished_species)
```
